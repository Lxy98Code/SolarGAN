# Purpose

This notebook shows how to plot the processed HMI and AIA 304 data from the paper "A Machine-learning Data Set Prepared from the NASA Solar Dynamics Observatory Mission" to closely match the images shown in the paper "Solar farside magnetograms from deep learning analysis of STEREO/EUVI data" as shown below:  

![](kim_et_al_2019.png)

The notebook makes extensive use of `sunpy` to ensure coordinate are handled correctly. The following function is used in the process of creating `sunpy.Map` objects from the processed data:

https://docs.sunpy.org/en/stable/generated/gallery/map/map_from_numpy_array.html#sphx-glr-generated-gallery-map-map-from-numpy-array-py

```python
%matplotlib inline
```

```python
import copy 

import numpy as np
import matplotlib.pyplot as plt

import astropy.units as u
from astropy.coordinates import SkyCoord

import sunpy.map
from sunpy.coordinates import frames
```

## HMI image

```python
hmi_npz_file = np.load('HMI20170901_1200_bx.npz')
hmi_20171101_1200 = hmi_npz_file['x']
hmi_npz_file.close()
```

```python
obstime = '2017-09-01 12:00'
observer = 'earth'  
arcsec_per_pixel = 4.8
rsun = 976  # in arcsec
```

```python
coord = SkyCoord(0*u.arcsec, 0*u.arcsec, obstime=obstime,
                 observer=observer, frame=frames.Helioprojective)
```

```python
header = sunpy.map.make_fitswcs_header(hmi_20171101_1200, coord,
                                       reference_pixel=None,
                                       scale=[arcsec_per_pixel, arcsec_per_pixel]*u.arcsec/u.pixel)
```

```python
header['rsun_obs'] = rsun
```

```python
hmi_map = sunpy.map.Map(hmi_20171101_1200, header)
```

```python
centers = sunpy.map.all_coordinates_from_map(hmi_map)
```

```python
interior = sunpy.map.coordinate_is_on_solar_disk(centers)
```

```python
hmi_map.data[~interior] = float('nan')
```

```python
hmi_cmap = copy.copy(hmi_map.cmap)
hmi_cmap.set_bad('k')
```

```python
fig = plt.figure(figsize=(10, 10))
hmi_map.plot(clip_interval=(0.05, 99.95)*u.percent, cmap=hmi_cmap)
plt.gca().grid(False)
plt.colorbar();
```

## AIA image

```python
aia_npz_file = np.load('AIA20170901_1200_0304.npz')
```

```python
aia_304_20171101_1200 = aia_npz_file['x']
```

```python
header = sunpy.map.make_fitswcs_header(aia_304_20171101_1200, coord,
                                       reference_pixel=None,
                                       scale=[arcsec_per_pixel, arcsec_per_pixel]*u.arcsec/u.pixel)
```

```python
header['rsun_obs'] = rsun
```

```python
aia_map = sunpy.map.Map(aia_304_20171101_1200, header)
```

```python
fig = plt.figure(figsize=(10, 10))
aia_map.plot(clip_interval=(0.25, 99.75)*u.percent)
plt.gca().grid(False)
plt.colorbar();
```

```python
aia_cmap = sunpy.visualization.colormaps.cm.sdoaia304
```

```python
fig = plt.figure(figsize=(10, 10))
aia_map.plot(clip_interval=(0.25, 99.75)*u.percent, cmap=aia_cmap)
plt.gca().grid(False)
plt.colorbar();
```
