## Clip to disk

```python
def all_corners_on_disk(smap):
    """
    Return a mask with True for those pixels of *smap* with all corners inside the solar disk.
    
    This could be done with:
        hpc_coords = all_coordinates_from_map(aia)
        mask = coordinate_is_on_solar_disk(hpc_coords)
        
    However, this considers pixel centers.
    """
    corners = sunpy.map.all_corner_coords_from_map(smap)
    corners_on_disk = sunpy.map.coordinate_is_on_solar_disk(corners)
    mask = np.empty_like(smap.data, dtype=bool)
    for i in range(smap.data.shape[0]):
        for j in range(smap.data.shape[1]):
            if all([corners_on_disk[i][j], corners_on_disk[i][j+1], corners_on_disk[i+1][j], corners_on_disk[i+1][j+1]]):
                mask[i][j] = True
            else:
                mask[i][j] = False
    return mask
```

```python
aia_mask = all_corners_on_disk(aia_map)
aia_masked_map = sunpy.map.Map(aia_map.data, aia_map.wcs, mask=~aia_mask)
```

```python
hmi_mask = all_corners_on_disk(hmi_map_rotated)
hmi_masked_map_rotated = sunpy.map.Map(hmi_map_rotated.data, hmi_map_rotated.wcs, mask=~hmi_mask)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(6, 12))
im = aia_masked_map.plot(axes=ax[0], clip_interval=(40, 99.9)*u.percent)
aia_masked_map.draw_limb()
fig.colorbar(im, ax=ax[0])

im = hmi_masked_map_rotated.plot(axes=ax[1], clip_interval=(1, 99.5)*u.percent)
hmi_masked_map_rotated.draw_limb()
fig.colorbar(im, ax=ax[1])
```

```python
output, footprint = reproject_interp(hmi_map_rotated, aia_masked_map.wcs, aia_masked_map.data.shape)
```

```python
hmi_masked_reproject_map = sunpy.map.Map(output, aia_masked_map.wcs, mask=aia_masked_map.mask)
```
